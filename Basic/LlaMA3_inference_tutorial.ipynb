````python
import os
import torch

from transformers import AutoTokenizer, AutoModelForCausalLM
````

````python
# HuggingFace에서 제공하는 모델과 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = AutoModelForCausalLM.from_pretrained("gpt2")

model.eval()
````

````python
model_id = 'beomi/Llama-3-Open-Ko-8B-Instruct-preview'

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype="auto",
    device_map="auto",
)
model.eval()
````

* `torch_dype`: 모델의 **가중치(weights)**를 **어떤 데이터 타입으로 로드**할지 지정. 기본적으로는 float32 형식을 사용하지만, 메모리 사용량을 줄이기 위해 다른 형식을 사용할 수 있음.
    * `auto` =>  일반적으로 float32나 float16을 자동 선택.

* `device_map`: **모델의 각 부분**을 **어떤 디바이스에 로드**할지 지정하는 파라미터. 주로 모델 병렬화(parallelization)와 관련이 있으며, 대형 모델을 여러 GPU에 분산하여 로드할 때 사용됨.
    * `auto`: 사용할 수 있는 디바이스에 따라 자동으로 모델의 각 부분을 할당.
    * `balanced`: 모델의 각 부분을 디바이스 메모리에 균형 있게 분산시킴.
    * `sequential`: 모델의 각 부분을 순차적으로 디바이스에 할당
    * 특정 디바이스에 모델의 특정 부분을 명시적으로 할당할 수 있음. ex) `{"transformer.layer.0": 0, "transformer.layer.1": 1}`는 첫 번째 레이어를 첫 번째 GPU에, 두 번째 레이어를 두 번째 GPU에 할당

## AutoTokenizer 예시

````python
# 입력 텍스트 설정
input_text = "Are you pissed off, John? Yeah~."

# 입력 텍스트를 토큰화 ->  PyTorch 텐서 형식으로 변환
input_ids = tokenizer.encode(input_text, return_tensors="pt")

print(len(input_text))
````

````python
print(input_ids.shape) # (Batch, Seq_Length)
````

* `Sequence Length`: 토큰화된 입력 문장의 길이

## Prompt x Inst x Msg

````python
PROMPT = '''당신은 유용한 AI 어시스턴트입니다. 사용자의 질의에 대해 친절하고 정확하게 답변해야 합니다.'''
INSTRUCTION = "다음 제목의 논문을 요약해줘 'Optimizing Language Augmentation for Multilingual Large Language Models: A Case Study on Korean'"

messages = [
    {"role": "system", "content": f"{PROMPT}"},
    {"role": "user", "content": f"{INSTRUCTION}"}
]
````

* `PROMPT`: 시스템 메시지로 사용될 내용. **AI 어시스턴트의 성격과 역할을 정의**. 즉, AI가 응답할 때 따라야 할 기본 지침을 설정.
* `INSTRUCTION`: **사용자 메시지**로, AI에게 특정 작업을 요청함.
* `messages` 리스트: **대화의 흐름을 정의**하는 메시지 목록. role과 content로 구성
    * `role`: 메시지 역할을 정의. system, user, assistant 등이 사용됨.
    * `content`: 해당 역할의 메시지 내용을 포함함.
    

**역할 설명**

System Message (`role: "system"`) :

* AI의 초기 지침을 설정함. 이 메시지는 대화의 맥락을 제공하며, AI 어시스턴트의 행동 방침을 정의함.
* 예시의 경우, AI에게 친절하고 정확하게 답변해야 함을 지시함.

User Message (`role: "user"`) :

* 사용자의 요청이나 질문을 포함. 이 메시지는 AI에게 특정 작업을 요청함.
* 예시는 사용자가 논문 제목을 제공하고, AI에게 이를 요약해달라고 요청함.

Assistant Message (`role: "assistant"`):
*  AI 응답을 나타내는 부분. AI 응답은 모델이 생성한 후에 추가될 수 있음.

## Tokenizer

````python
input_ids = tokenizer.apply_chat_template(
    messages,
    add_generation_prompt=True,
    return_tensors="pt"
).to(model.device)
````

`tokenizer.apply_chat_template`:
* 주어진 `messages` 목록을 모델이 이해할 수 있는 형식으로 변환함. 이 메서드는 주로 대화형 AI 모델에서 사용됨.

`messages`:
* 대화의 맥락을 포함하는 메시지 목록. 이 목록은 **시스템 메시지**, **사용자 메시지**, 그리고 **AI 응답**을 포함.


`add_generation_prompt=True`:
* 모델이 응답을 생성할 때 추가적인 프롬프트를 삽입하여 더 나은 결과를 유도하도록 함. ex) `
"AI: "`와 같은 프롬프트를 추가할 수 있음.

## Terminator

````python
terminators = [
    tokenizer.eos_token_id,
    tokenizer.convert_tokens_to_ids("<|eot_id|>")
]
````

이 코드는 두 개의 **종료 토큰(terminator tokens)**을 정의하여, 모델이 텍스트 생성을 멈출 조건을 설정하는 데 사용됨.

(ref) [Please add <|eot_id|> as a stop token to the HF config, meta-llama.](https://huggingface.co/meta-llama/Meta-Llama-3-70B-Instruct/discussions/12)

<br>

`tokenizer.eos_token_id`:
* "end of sequence(eos)" 토큰의 ID. 이 토큰은 **모델이 텍스트 생성을 완료했음을 나타내기** 위해 사용됨.
* 모델이 이 토큰을 생성하면 더 이상 텍스트를 생성하지 않음.

`tokenizer.convert_tokens_to_ids("<|eot_id|>")`
* 이 메서드는 **주어진 토큰을 해당하는 토큰 ID로 변환**함.  이 경우, "end of text(eot)" 토큰의 ID `"<|eot_id|>"`로 변환함.

````python
print(tokenizer.eos_token_id)
print(tokenizer.decode(tokenizer.eos_token_id, skip_special_tokens=False))
````

````python
print(tokenizer.convert_tokens_to_ids("<|eot_id|>"))
print(tokenizer.decode(tokenizer.convert_tokens_to_ids("<|eot_id|>"), skip_special_tokens=False))
````

## Generate

````python
outputs = model.generate(
    input_ids,
    max_new_tokens=512,
    eos_token_id=terminators,
    do_sample=True,
    temperature=1,
    top_p=0.9,
)
````

`max_new_tokens`:
* 모델이 새로 생성할 토큰의 최대 개수

`eos_token`:
* 모델이 텍스트 생성을 종료할 때 사용할 종료 토큰의 ID를 지정

`do_sample`:
* `True` 이면, 모델이 다음 토큰을 샘플링 방식으로 생성함. 샘플링 방식은 확률 분포를 기반으로 하여 더 다양하고 창의적인 텍스트 생성이 가능.
* `False` 이면, 모델은 그리디 서치(greedy search) 또는 빔 서치(beam search)와 같은 결정론적 방법을 사용.

`temperature`:
* 샘플링 과정에서 사용되는 온도 매개변수. 높을수록 모델은 더 무작위적이고 다양하게 선택함.
* 낮은 값(0.1에 가까운 값)은 더 결정론적인 출력을, 높은 값(1.0 이상)은 더 무작위적인 출력을 생성합니다.

`top_p`:
* Nucleus 샘플링(Top-p 샘플링)에서 사용되는 파라미터.
* **더 다양한 출력**(높은 top_p 값)과 **더 집중된 출력**(낮은 top_p 값) 사이의 균형을 조절합니다.
* top_p=0.9는 확률 분포의 상위 90%에 속하는 토큰들만 고려하여 샘플링

````python
outputs.shape  # (Batch_size, Seq_length)
````

* `Batch_size:=1`: 한 번의 요청으로 하나의 시퀀스가 생성
* `Seq_length`: 모델이 생성한 총 토큰 수. 여기에는 초기 입력 텍스트의 토큰과 새로 생성된 토큰이 모두 포함됨.
  * 즉, Seq_length = 초기 입력 토큰 수 + 생성된 새로운 토큰 수

# Chatbot Interface
위의 단발성 문장 생성을 활용해서 Chatbot 만들기.

````python
import gradio as gr
````

````python
PROMPT = '''당신은 유용한 AI 어시스턴트입니다. 사용자의 질의에 대해 친절하고 정확하게 답변해야 합니다.'''
````

````python
def generate_response(input_text, chat_history):
    messages = [{"role": "system", "content": f"{PROMPT}"}]

    for message in chat_history:
        messages.append({"role": "user", "content": message[0]})
        messages.append({"role": "assistant", "content": message[1]})  # AI assistant 추가

    messages.append({"role": "user", "content": input_text})

    # -- 입력 토큰 생성
    input_ids = tokenizer.apply_chat_template(
        messages,
        add_generation_prompt=True,
        return_tensors="pt"
    ).to(model.device)

    # -- 터미네이터 리스트
    terminators = [
        tokenizer.eos_token_id,
        tokenizer.convert_tokens_to_ids("<|eot_id|>")
    ]

    # -- 토큰 생성
    outputs = model.generate(
        input_ids,
        max_new_tokens=512,
        eos_token_id=terminators,
        do_sample=True,
        temperature=1,
        top_p=0.9,
    )

    # -- 입력 토큰 빼고 새롭게 생성된 토큰만 출력
    response = outputs[0][input_ids.shape[-1]:]
    response = tokenizer.decode(response, skip_special_tokens=True)

    chat_history.append((input_text, response))

    return "", chat_history
````

````python
with gr.Blocks() as demo:
    chatbot = gr.Chatbot(height=600)
    msg = gr.Textbox()
    clear = gr.Button("Clear")

    msg.submit(generate_response, [msg, chatbot], [msg, chatbot])
    clear.click(lambda: None, None, chatbot, queue=False)

demo.launch()
````
